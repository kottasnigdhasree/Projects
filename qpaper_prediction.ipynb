```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
q_vect = vectorizer.fit_transform(qpaper['Q'])
```

```python
print(vectorizer.get_feature_names())
```

```python
counts = pd.DataFrame(q_vect.toarray(),
columns=vectorizer.get_feature_names())
```

```python
counts
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = ["are", "and", "the", "with","a","an", "its", "what", "why", "explain", "is", "in" ]

word_tokens = vectorizer.get_feature_names()

filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

filtered_sentence = []

for w in word_tokens:
	if w not in stop_words:
		filtered_sentence.append(w)

print(word_tokens)
print(filtered_sentence)
```

```python
remv = set(word_tokens)-set(filtered_sentence)
remv = list(remv)
```

```python
count1 = counts.drop(remv,axis = 1)
```

```python
count1 = pd.DataFrame(count1.T.sort_values(by=0, ascending=False))
count1.head()
```
